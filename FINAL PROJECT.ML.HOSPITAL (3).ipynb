```python
import pandas as pd
```

This project uses machine learning to predict hospital resource utilization efficiency based on patient and hospital system data. The goal is to help hospitals allocate resources more effectively by identifying key factors that influence efficient usage.

# LABEL ENCODING

```python
from sklearn.preprocessing import LabelEncoder
```

```python
j=LabelEncoder()
k=j.fit_transform(a['Patient_ID'])
l=j.fit_transform(a['Gender'])
m=j.fit_transform(a['Health_Status'])
n=j.fit_transform(a['Disease_Type'])
o=j.fit_transform(a['Resource_Needed'])
p=j.fit_transform(a['Sensor_ID'])
q=j.fit_transform(a['Resource_ID'])
a['Patient_ID']=k
a['Gender']=l
a['Health_Status']=m
a['Disease_Type']=n
a['Resource_Needed']=o
a['Sensor_ID']=p
a['Resource_ID']=q
a
```

# CORRELATION

```python
a.corr()
```

```python
b=a.drop(['Patient_ID','Sensor_ID','Resource_ID'],axis='columns')
b
```

# TRAIN_TEST_SPLIT

```python
from sklearn.model_selection import train_test_split
```

```python
x=b.drop('Resource_Utilization_Efficiency',axis='columns')
x
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=45)
xtrain,xtest,ytrain,ytest
```

# VISUALIZATION

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Check correlation of all features and the target
plt.figure(figsize=(10,8))
sns.heatmap(b.corr(), annot=True, cmap='coolwarm')
plt.title("Feature Correlation Heatmap")
plt.show()
```

The heatmap shows the correlation between features and the target. It helps identify which variables are strongly related to the target. Features with higher positive or negative correlation are more useful for prediction. This visualization guided the feature selection for model building.



```python
sns.countplot(x='Resource_Utilization_Efficiency', data=b, palette='Set2')
plt.title('Target Variable Distribution')
plt.xlabel('Resource Allocation (0 = No, 1 = Yes)')
plt.ylabel('Count')
plt.show()
```

This bar plot shows how many records belong to each class of the target variable Resource_Utilization_Efficiency. It helps check if the dataset is balanced or imbalanced. A balanced dataset ensures the model learns fairly for both classes.

```python
sns.boxplot(x='Resource_Utilization_Efficiency', y='Age', data=b, palette='pastel')
plt.title('Age Distribution by Resource Utilization Efficiency')
plt.xlabel('Resource Utilization Efficiency (0 = No, 1 = Yes)')
plt.ylabel('Age')
plt.show()
```

The boxplot compares patient age across the two target classes. It helps identify if age influences resource utilization efficiency. Differences in the age distributions suggest age may be an important factor.

```python
sns.countplot(x='Gender', hue='Resource_Utilization_Efficiency', data=b, palette='husl')
plt.title('Resource Utilization by Gender')
plt.xlabel('Gender (Encoded)')
plt.ylabel('Count')
plt.legend(title='Resource Utilization Efficiency')
plt.show()
```

This bar plot shows how resource utilization efficiency is distributed between genders. It helps reveal if one gender receives or utilizes hospital resources more efficiently than the other.
